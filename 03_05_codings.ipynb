```python
import os
import pandas as pd
from bs4 import BeautifulSoup
import textstat
```

```python
agents_file = 'agents.tsv'
agents_df = pd.read_csv(os.path.join('input', 'codings', agents_file), sep='\t')

task_descriptions_file = 'task_descriptions.tsv'
task_descriptions_df = pd.read_csv(os.path.join('input', 'codings', task_descriptions_file), sep='\t')

comments_file = 'comments.tsv'
comments_df = pd.read_csv(os.path.join('input', 'codings', comments_file), sep='\t')
```

```python
display(agents_df.head(20))
display(task_descriptions_df.head(20))
display(comments_df.head(20))
```

## Agents

```python
co_creative_roles = agents_df["co_creative_role"].str.split(", ", regex=False, expand=True)
display(co_creative_roles)

co_creative_roles_flat = pd.Series([])
for series_name, series in co_creative_roles.items():
    co_creative_roles_flat = pd.concat([co_creative_roles_flat, series[series.notna()]])

display(len(co_creative_roles_flat))
display(co_creative_roles_flat.value_counts())
```

## Task descriptions

```python
desc_writing_stages = task_descriptions_df["writing_stage"].str.split(", ", regex=False, expand=True)
display(desc_writing_stages)

desc_writing_stages_flat = pd.Series([])
for series_name, series in desc_writing_stages.items():
    desc_writing_stages_flat = pd.concat([desc_writing_stages_flat, series[series.notna()]])

display(len(desc_writing_stages_flat))
display(desc_writing_stages_flat.value_counts())
```

```python
desc_intent = task_descriptions_df["users_primary_intent"].str.split(", ", regex=False, expand=True)
display(desc_intent)

desc_intent_flat = pd.Series([])
for series_name, series in desc_intent.items():
    desc_intent_flat = pd.concat([desc_intent_flat, series[series.notna()]])

display(len(desc_intent_flat))
display(desc_intent_flat.value_counts())
```

```python
task_descriptions = task_descriptions_df["description"]
display(task_descriptions)

parsed_text_list = []
word_counts = []
for idx, text in task_descriptions.items():
    parsed_text_list.append(BeautifulSoup(str(text), 'lxml').get_text())
    word_counts.append(textstat.lexicon_count(text, removepunct=True))

display(pd.Series(word_counts).describe())
```

```python
display(task_descriptions_df.tail(10))
```

```python
task_descriptions_df_copy = task_descriptions_df.copy()

task_descriptions_df_copy.assignee_ids = task_descriptions_df_copy.assignee_ids.str.split(',')
task_descriptions_df_copy = task_descriptions_df_copy.explode('assignee_ids')
display(task_descriptions_df_copy.tail(15))
```

```python
descriptions_agents_df = pd.merge(task_descriptions_df_copy, agents_df, left_on='assignee_ids', right_on='id', how='left')
display(descriptions_agents_df.head(15))
```

```python
description_agents_exploded_df = descriptions_agents_df.copy()
display(description_agents_exploded_df.head(15))

description_agents_exploded_df.co_creative_role = description_agents_exploded_df.co_creative_role.str.split(',')
description_agents_exploded_df = description_agents_exploded_df.explode('co_creative_role')
description_agents_exploded_df.co_creative_role = description_agents_exploded_df.co_creative_role.str.strip()
description_agents_exploded_df.co_creative_role = description_agents_exploded_df.co_creative_role.fillna("ai-author")
display(description_agents_exploded_df.head(15))

display("Co-Creative-Roles for Tasks in OUTLINING stage")
outlining_roles = description_agents_exploded_df[description_agents_exploded_df['writing_stage'] == "outlining"]["co_creative_role"]
display(outlining_roles.value_counts())
display(outlining_roles.value_counts().describe())

display("Co-Creative-Roles for Tasks in DRAFTING stage")
drafting_roles = description_agents_exploded_df[description_agents_exploded_df['writing_stage'] == "drafting"]["co_creative_role"]
display(drafting_roles.value_counts())
display(drafting_roles.value_counts().describe())

display("Co-Creative-Roles for Tasks in REVISING stage")
revising_roles = description_agents_exploded_df[description_agents_exploded_df['writing_stage'] == "revising"]["co_creative_role"]
display(revising_roles.value_counts())
display(revising_roles.value_counts().describe())
```

```python
comments_writing_stages = comments_df["writing_stage"].str.split(", ", regex=False, expand=True)
display(comments_writing_stages)

comments_writing_stages_flat = pd.Series([])
for series_name, series in comments_writing_stages.items():
    comments_writing_stages_flat = pd.concat([comments_writing_stages_flat, series[series.notna()]])

display(len(comments_writing_stages_flat))
display(comments_writing_stages_flat.value_counts())
```

```python
comments_intent = comments_df["users_primary_intent"].str.split(", ", regex=False, expand=True)
display(comments_intent)

comments_intent_flat = pd.Series([])
for series_name, series in comments_intent.items():
    comments_intent_flat = pd.concat([comments_intent_flat, series[series.notna()]])

display(len(comments_intent_flat))
display(comments_intent_flat.value_counts())
```

```python
comments = comments_df["text"]
display(comments)
display(len(comments))

ai_mentionings = [
    "@aiauthor",
    "@aiAI Max",
    "@aiTextStructurer",
    "@aiErika",
    "@aiContentReviewer",
    "@aiBest HCI Professor",
    "@aiTextReviewer",
    "@aiInteraction Designer",
    "@aiThe AI Expert",
    "@aiFun facts",
    "@aiAiDevelop",
    "@aiRephraser",
    "@aiGeorge",
    "@aiSynthesizer",
    "@aiProposal Brain",
    "@aiGrammar",
    "@aiFact"
]

parsed_text_list = []
word_counts = []
for idx, text in comments.items():
    t = text
    for mention in ai_mentionings:
        if mention in t:
            t = t.replace(mention, "")
            
    parsed_text_list.append(BeautifulSoup(str(t), 'lxml').get_text())
    word_counts.append(textstat.lexicon_count(t, removepunct=True))

# print("parsed_text_list")
# display(len(parsed_text_list))
# display(parsed_text_list)
# 
# print("comments raw")
# display(comments.to_list())

display(parsed_text_list)
display(word_counts)
display(pd.Series(word_counts).describe())
```
